Student performance according to practice in sports

```python
def plot_histograms(df, group_by, score_columns):
    unique_groups = df[group_by].unique()
    n_groups = len(unique_groups)
    n_scores = len(score_columns)
    
    fig, axes = plt.subplots(n_scores, n_groups, figsize=(15, 5 * n_scores), sharey=True)
    
    for i, score in enumerate(score_columns):
        for j, group in enumerate(unique_groups):
            subset = df[df[group_by] == group]
            axes[i, j].hist(subset[score], bins=10, alpha=0.7)
            axes[i, j].set_title(f'{score} for {group}')
            axes[i, j].set_xlabel('Scores')
            axes[i, j].set_ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
```

```python

# Plot histograms for PracticeSport
plot_histograms(dataset, 'PracticeSport', ['MathScore', 'ReadingScore', 'WritingScore'])
```

```python
# Plot histograms for TransportMeans
plot_histograms(dataset, 'TransportMeans', ['MathScore', 'ReadingScore', 'WritingScore'])
```
